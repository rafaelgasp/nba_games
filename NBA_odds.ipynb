```python
def trata_df_odds(table_html):
    df_odds = pd.read_html(table_html)
    df_odds = pd.concat(df_odds)
    df_odds[0] = df_odds[0].shift(1)
    
    df_odds.columns = ["date", "game", "result", "odds_home", "odds_away", "B's"]
    df_odds.dropna(subset=(["odds_home", "B's"]), inplace=True)
    
    df_odds["date"] = [x if len(x) > 5 else np.nan for x in df_odds["date"]]
    df_odds["date"].fillna(method='ffill', inplace=True)
    
    return(df_odds)
```

```python
list_odds = []

for season in seasons:    
    for page in range(1, 50):
        print(season, page)
        driver.get("http://www.oddsportal.com/basketball/usa/nba-" + season + "/results/#/page/" + str(page) + "/")
        time.sleep(5)
        
        tabela = driver.find_element_by_id("tournamentTable").get_attribute('innerHTML')
        
        if "No data available" in tabela:
            #print(tabela)
            break
        else:
            df_odds = trata_df_odds(tabela)
            list_odds.append(df_odds)        
```

```python
base_odds = pd.concat(list_odds).reset_index(drop=True)
base_odds.drop_duplicates(inplace=True)
```

```python
base_odds["fl_playoff"] = [1 if "Play Offs" in x else 0 for x in base_odds["date"]]
base_odds["fl_pre_season"] = [1 if "Pre-season" in x else 0 for x in base_odds["date"]]
```

```python
base_odds["data"] = [datetime.strptime(x.split(" - ")[0], '%d %b %Y') for x in base_odds["date"]]
```

```python
def parse_home_team(game_str):
    for key in de_para_siglas.keys():
        game_str = game_str.replace(key, de_para_siglas[key])
    
    game_str = game_str.replace("-", "@")
    spl =  game_str.split(" @ ")
    return(game_str, spl[0], spl[1])
```

```python
base_odds["team_home"] = [parse_home_team(x)[2] for x in base_odds["game"]]
base_odds["team_away"] = [parse_home_team(x)[1] for x in base_odds["game"]]

base_odds["match"] = [parse_home_team(x)[0] for x in base_odds["game"]]
base_odds["match"] = base_odds["match"] + " " + base_odds["data"].map(lambda x: x.strftime('%Y-%m-%d'))

base_odds["PTS_home"] = [int(x.split(":")[0]) for x in base_odds["result"]]
base_odds["PTS_away"] = [int(x.split(":")[1].split()[0]) for x in base_odds["result"]]
```

```python
base_odds['fl_home_win'] = np.where(base_odds['PTS_home'] > base_odds['PTS_away'], 1, 0)
```

```python
base_odds.columns = ['date_ref', 'game_full_names', 'result', 'odds_away', 'odds_home', "B's", 'fl_playoff',
       'fl_pre_season', 'date', 'team_home', 'team_away', 'match', 'PTS_home',
       'PTS_away', 'fl_home_win']
```

```python
base_odds.set_index("match", inplace=True)
```
