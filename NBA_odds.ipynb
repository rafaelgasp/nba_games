```python
import selenium
import pandas as pd
import numpy as np
import os
from datetime import datetime
from selenium import webdriver
```

```python
driver = webdriver.Chrome("C:/Users/rafae/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
driver.get("http://www.oddsportal.com/basketball/usa/nba-2017-2018/results/page/1/")
```

```python
df_odds = pd.read_html(driver.find_element_by_id("tournamentTable").get_attribute('innerHTML'))
df_odds = pd.concat(df_odds)
```

```python
df_odds[0] = df_odds[0].shift(1)
```

```python
df_odds.columns = ["date", "game", "result", "odds_away", "odds_home", "B's"]
df_odds.dropna(subset=(["odds_home", "B's"]), inplace=True)
```

```python
df_odds["date"] = [x if len(x) > 5 else np.nan for x in df_odds["date"]]
```

```python
df_odds["date"].fillna(method='ffill', inplace=True)
```
